```python
def get_day(datetime_str: datetime, date_format: str =  '%Y-%m-%dT%H:%M:%S') -> int:
    from datetime import datetime
    _datetime_str = datetime_str
    datetime_object = datetime.strptime(_datetime_str, '%Y-%m-%dT%H:%M:%S')
    return datetime_object.day
    
def get_month(datetime_str: datetime, date_format: str =  '%Y-%m-%dT%H:%M:%S') -> int:
    from datetime import datetime
    _datetime_str = datetime_str
    datetime_object = datetime.strptime(_datetime_str, '%Y-%m-%dT%H:%M:%S')
    return datetime_object.month
    
def get_year(datetime_str: datetime, date_format: str =  '%Y-%m-%dT%H:%M:%S') -> int:
    from datetime import datetime
    _datetime_str = datetime_str
    datetime_object = datetime.strptime(_datetime_str, '%Y-%m-%dT%H:%M:%S')
    return datetime_object.year
```

```python
dato = '2020-07-04T17:00:00'
get_year(dato)
get_day(dato)
get_month(dato)
```
